# Visualization of smooth functions

This notebook visualizes some smooth functions used in [Learning To Solve Circuit-SAT: An Unsupervised Differentiable Approach](https://openreview.net/pdf?id=BJxgz2R9t7).

```python
import math
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

## smooth Step function

```python
def L(s, k):
    a = (1 - s)**k
    b = s**k
    return a / (a+b)
```

```python
s = np.linspace(0,1,100)

for k in [1,2,5,10,20,100]:
    plt.plot(s, L(s, k=k), label=f"k={k}")
    plt.legend()
```

## smooth min/max function

```python
def smooth_max(*xs, t=1):
    return sum(a * math.exp(a / t) for a in xs) / sum(math.exp(a / t) for a in xs)

def smooth_min(*xs, t=1):
    return sum(a * math.exp(- a / t) for a in xs) / sum(math.exp(- a / t) for a in xs)
```

```python
t = 10**np.linspace(-2,3,1000)
```

```python
plt.xlabel("t")
plt.semilogx(t, [smooth_min(0.2, 0.5, t=x) for x in t], label="min(0.2, 0.5)")
plt.semilogx(t, [smooth_min(0.2, 0.8, t=x) for x in t], label="min(0.2, 0.8)")
plt.semilogx(t, [smooth_min(0.4, 0.6, t=x) for x in t], label="min(0.4, 0.6)")
plt.legend()
```

```python
plt.xlabel("t")
plt.semilogx(t, [smooth_max(0.2, 0.5, t=x) for x in t], label="max(0.2, 0.5)")
plt.semilogx(t, [smooth_max(0.2, 0.8, t=x) for x in t], label="max(0.2, 0.8)")
plt.semilogx(t, [smooth_max(0.4, 0.6, t=x) for x in t], label="max(0.4, 0.6)")
plt.legend()
```
